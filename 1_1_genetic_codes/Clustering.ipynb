To visualize the output for different sizes of words we encapsulate these methods

### Helper functions

```python
def read_genomes(filename, genome_size):
    """read the full genome string and split the full pattern in sub strings of size genome_size"""
    
    with open(filename) as file:
        genomes = file.readlines()
        
    genomes = [genome.strip() for genome in genomes]
    genomes = "".join(genomes)
    
    genomes_list = []
    
    for i in range(0, len(genomes), genome_size):
        genomes_list.append(genomes[i:i + genome_size])
        
    return genomes_list


def get_frequencies(genomes, word_size):
    """get the frequencies of the words of size word_size on the genomes patterns"""
    
    words = list(itertools.product('agct', repeat = word_size))
    words = ["".join(item) for item in words]

    total_frequencies = []

    for genome in genomes:
        frequencies = []
        for word in words:
            frequencies.append(genome.count(word))
        total_frequencies.append(frequencies)
        
    return total_frequencies


def get_pca_features(total_frequencies, pca_components):
    """get the standarized pca features of the frequencies considering a number of components pca_components"""
    
    pca = PCA(n_components = pca_components)
    pca.fit(total_frequencies)
    pca_features = pca.transform(total_frequencies)
    pca_features = StandardScaler().fit_transform(pca_features)
    
    return pca_features
        
```

```python
genomes = read_genomes("ccrescentus.txt", 300)

for word_size in range(1,5):
    frequencies = get_frequencies(genomes, word_size)
    pca_features = get_pca_features(frequencies, 2)
    plt.scatter(pca_features[:,0], pca_features[:,1], s = 2)
    plt.show()
```

# 3. Clustering and visualizing results

We apply clustering to the best output of the previous step

```python
frequencies = get_frequencies(genomes, 3)
pca_features = get_pca_features(frequencies, 2)
```

```python
model = KMeans(n_clusters = 7)
model.fit(pca_features)
```

```python
n_clusters = model.predict(pca_features)
n_clusters
```

```python
plt.scatter(pca_features[:,0], pca_features[:,1], c = n_clusters, s = 3)
```
